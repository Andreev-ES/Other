Реализовать прототип консольной программы - проводника, для работы с файлами. Создать функции создания, удаления, перемещения, копирования(файла, папки) с использованием системы контроля версий git.
```python
import os
import shutil

# Функция вызывающая основное меню :
def main_menu():
    menu_item = input(
                    'Создать файл - 1:\n'
                    'Создать папку - 2:\n'
                    'Удалить файл  - 3\n'
                    'Удалить папку  - 4\n'
                    'Переместить файл или папку - 5\n'
                    'Скопировать - 6\n'
                    'Для выхода из программы введите-Q\n'
                )
    menu_item = menu_item.upper()
    return menu_item

#Функция для возврата пути
def get_path(description):
    path = input(f'{description}: \n')
    return path

#Функция для проверки наличия файла папки
def check_exists(path, type_check):
    if type_check == 'folder':
        check = os.path.isdir(path)
    if type_check == 'file':
        check = os.path.isfile(path)
    return check

#Функция для создания файла
def create_file(path):
    try:
        if check_exists(path,'file'):
            query = input('\nТакой файл уже существует, перезаписать?(y/n)\n')
            if query.lower() == 'y':
                open(path, 'w')
                print(f'\nФайл перезаписан:{path}')
        else:
            open(path, 'w')
            print(f'\nФайл создан:{path}')
    except Exception as err:
        print(err) 
        
#Функция для создания папки        
def create_folder(path):
    if check_exists(path,'folder'):
        print('Такая папка уже существует!\n')
    else:
        os.makedirs(path)
        print(f'\nПапка создана:{path}')
        
#Функция для удаления файла        
def remove_file(path):
    if check_exists(path,'file'):
        os.remove(path)
        print(f'\nФайл удален:{path}')
    else:
        print(f'\nФайл по указанному пути не существует!')

#Функция для удаления папки        
def remove_folder(path):
    if check_exists(path,'folder'):
        os.rmdir(path)
        print(f'\nПапка удалена:{path}')
    else:
        print(f'\nПапки по указанному пути не существует!') 
        
#Функция для перемещения файла или папки
def move_file_folder(src_path, trg_path):
    try:
        shutil.move(src_path, trg_path)
    except Exception as err:
        print(err)
        
#Функция для копирования файла или папки
def copy_file_folder(src_path, trg_path):
    try:
        shutil.copy(src_path, trg_path)
    except Exception as err:
        print(err)           
```

```python
qiut = ''
while qiut != 'Q':
    menu_item = main_menu()
    if menu_item.upper() == 'Q':
        qiut = menu_item
    elif menu_item == '1':
        path = get_path('Укажите место расположения файла')
        create_file(path)
    elif menu_item == '2':
        path = get_path('Укажите место расположения папки')
        create_folder(path)
    elif menu_item == '3':
        path = get_path('Укажите место расположения файла')
        remove_file(path) 
    elif menu_item == '4':
        path = get_path('Укажите место расположения папки')
        remove_folder(path)
    elif menu_item == '5':
            src_path = get_path('Укажите источник')
            trg_path = get_path('Укажите конечное месторасположение')
            move_file_folder(src_path, trg_path) 
    elif menu_item == '6':
            src_path = get_path('Укажите источник')
            trg_path = get_path('Укажите конечное месторасположение')
            copy_file_folder(src_path, trg_path)             
```
